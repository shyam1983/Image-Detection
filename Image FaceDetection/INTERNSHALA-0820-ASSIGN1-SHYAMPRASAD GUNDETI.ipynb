```python
import face_recognition

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
import numpy as np
import cv2
%matplotlib inline
```

```python
image1 = cv2.imread("images.jpg")
hrithik = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

image2 = cv2.imread("images1.jpg")
sharukh = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

image3 = cv2.imread("images2.jpg")
salman = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)
```

```python
hrithik_encoding = face_recognition.face_encodings(hrithik)[0]
sharukh_encoding = face_recognition.face_encodings(sharukh)[0]
salman_encoding = face_recognition.face_encodings(salman)[0]

known_face_encodings = [hrithik_encoding,sharukh_encoding,salman_encoding]
```

```python
#check by typing images.jpg for hrithik,  images1.jpg for sharukh, images2.jpg for salman, and women.jpg for unknown
IMAGES = input("check by typing images.jpg for hrithik,  images1.jpg for sharukh, images2.jpg for salman, and women.jpg for unknown:")
image4 = cv2.imread(IMAGES)
unknown_image = cv2.cvtColor(image4,cv2.COLOR_BGR2RGB)
plt.imshow(unknown_image)

unknown_face_encodings = face_recognition.face_encodings(unknown_image)
```

```python
from scipy.spatial import distance

for unknown_face_encoding in unknown_face_encodings:
    results = []
    for known_face_encoding in known_face_encodings:
        d = distance.euclidean(known_face_encoding, unknown_face_encoding)
        results.append(d)
    threshold=0.5
    results = np.array(results) <= threshold
    name = "unknown"
    
    if results[0]:
        name = "hrithik"
    elif results[1]: 
        name = "sharukh"
    elif results[2]:
        name = "salman"
    print(f"found {name} in the photo!")    
```
